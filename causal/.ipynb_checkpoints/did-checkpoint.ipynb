### 2x2 Design

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data/minwage.csv'
data = pd.read_csv(file_path)
print(data.columns.tolist())

location_wage_data = data.groupby('location')[['wageBefore', 'wageAfter']].mean()
data['location'] = data['location'].replace(r'^NJ.*', 'NJ', regex=True)
location_wage_combined = data.groupby('location')[['wageBefore', 'wageAfter']].mean()

unique_locations = data['location'].unique()
print(unique_locations)

data['location'] = data['location'].replace(['centralNJ', 'northNJ', 'shoreNJ', 'southNJ'], 'NJ')
combined_locations = data['location'].unique()
print(combined_locations)

for loc in data['location'].unique():
    subset = data[data['location'] == loc]
    plt.figure(figsize=(8, 5))
    subset['wageBefore'].plot(kind='kde', label='Wage Before', linewidth=2)
    subset['wageAfter'].plot(kind='kde', label='Wage After', linewidth=2)
    plt.title(f'KDE of Wages Before and After ({loc})')
    plt.xlabel('Wage')
    plt.ylabel('Density')
    plt.legend()
    plt.tight_layout()
    plt.show()

data['empBefore'] = data['fullBefore'] + data['partBefore']
data['empAfter'] = data['fullAfter'] + data['partAfter']
before_data = data[['location', 'empBefore']].rename(columns={'empBefore': 'emp'})
before_data['period'] = 0
after_data = data[['location', 'empAfter']].rename(columns={'empAfter': 'emp'})
after_data['period'] = 1
combined_data = pd.concat([before_data, after_data], ignore_index=True)
combined_data['location'] = combined_data['location'].apply(lambda x: 1 if x == 'NJ' else 0)
display(combined_data)
```

```python
import statsmodels.formula.api as smf
combined_data['location_period'] = combined_data['location'] * combined_data['period']
model = smf.ols('emp ~ location + period + location_period', data=combined_data).fit(cov_type='cluster', cov_kwds={'groups': combined_data['location']})
print(model.summary())
```

```python
import numpy as np

def plot_did_expanded(model, pre_date, post_date, title, post='period', treated='location', post_treated='location_period', y_min=None, y_max=None):
    alpha = model.params['Intercept']
    beta = model.params[post]
    gamma = model.params[treated]
    delta = model.params[post_treated]

    time_labels = [pre_date, post_date]
    pre_time = np.array([0])
    post_time = np.array([1])
    time = np.concatenate((pre_time, post_time))

    control_pre = alpha
    control_post = alpha + beta
    treatment_pre = alpha + gamma
    treatment_post = alpha + beta + gamma + delta
    counterfactual_post = alpha + beta + gamma

    control = np.array([control_pre, control_post])
    treatment = np.array([treatment_pre, treatment_post])
    counterfactual = np.array([treatment_pre, counterfactual_post])

    plt.figure(figsize=(8, 6))
    plt.plot(time, control, 'b-', marker='o', label='PA (Control)', linewidth=2)
    plt.plot(time, treatment, 'r-', marker='o', label='NJ (Treatment)', linewidth=2)
    plt.plot(time, counterfactual, 'r--', marker='o', label='NJ Counterfactual (No Treatment)', linewidth=2)
    plt.axvline(x=0.5, color='gray', linestyle='--', linewidth=1.5, label='Policy Change')

    plt.fill_between(post_time, treatment_post, counterfactual_post, alpha=0.2, color='red', label='Treatment Effect')

    plt.xlabel('Time', fontsize=14)
    plt.ylabel('Employment (Mean)', fontsize=14)
    plt.title(title, fontsize=16)
    plt.xticks([0, 1], time_labels, fontsize=14)
    plt.yticks(fontsize=14)
    
    if y_min is not None:
        plt.ylim(y_min, y_max)
        
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.legend(fontsize=14)
    plt.tight_layout()
    plt.show()

plot_did_expanded(model, 'Before', 'After', 'Difference-in-Differences: Expanded View with Counterfactual', y_min=20, y_max=35)
```

```python
# Compute means for the example table
mean_table = combined_data.groupby(['location', 'period'])['emp'].mean().unstack()

# Calculate differences for each group and the overall difference-in-differences
nj_diff = mean_table.loc[1, 1] - mean_table.loc[1, 0]  # NJ difference (Post - Pre)
pa_diff = mean_table.loc[0, 1] - mean_table.loc[0, 0]  # PA difference (Post - Pre)
did = nj_diff - pa_diff

# Construct the 2x2 table
result_table = pd.DataFrame({
    'Pre': [mean_table.loc[1, 0], mean_table.loc[0, 0], nj_diff - pa_diff],
    'Post': [mean_table.loc[1, 1], mean_table.loc[0, 1], nj_diff],
    'Difference': [nj_diff, pa_diff, did]
}, index=['New Jersey', 'Pennsylvania', 'Difference'])

# Display the result table
display(result_table)
```

### Multiple Time Periods

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data/WDI.dta'
treated_file_path = 'data/Treated.dta'

data = pd.read_stata(file_path)
print(data.columns)
treated_data = pd.read_stata(treated_file_path)

data['after'] = (data['year'] >= 2009).astype(int)
combined_data = data.merge(treated_data, on='country', how='left')
combined_data['treated'] = combined_data['treated'].fillna(0).astype(int)
combined_data['T_it'] = ((combined_data['treated'] == 1) & (combined_data['after'] == 1)).astype(int)

prepared_data = combined_data.rename(columns={'year': 't', 'country': 'i', 'gdppc': 'Y_it'})
prepared_data = prepared_data[['t', 'i', 'Y_it', 'T_it'] + [col for col in combined_data.columns if col not in ['year', 'country', 'gdppc', 'treated', 'after', 'T_it']]]
prepared_data = prepared_data.merge(combined_data[['year', 'country', 'treated']], left_on=['t', 'i'], right_on=['year', 'country'], how='left')
prepared_data = prepared_data.drop(columns=['year', 'country'])
prepared_data['treated_group'] = prepared_data['treated'].apply(lambda x: 'Treated Countries' if x == 1 else 'Untreated Countries')
avg_yit_grouped = prepared_data.groupby(['t', 'treated_group'])['Y_it'].mean().reset_index()

plt.figure(figsize=(10, 6))
for group, group_data in avg_yit_grouped.groupby('treated_group'):
    plt.plot(group_data['t'], group_data['Y_it'], label=group)
plt.title('Average GDP per Capita (Y_it) Over Time by Treated/Untreated Countries')
plt.xlabel('Year')
plt.ylabel('Average GDP per Capita')
plt.legend()
plt.grid(True)
plt.show()
```

#### TWFE - with and without clustered errors

```python
import pandas as pd
from linearmodels.panel import PanelOLS, compare

# Load data
file_path = 'data/WDI.dta'
treated_file_path = 'data/Treated.dta'

data = pd.read_stata(file_path)
print(data.columns)
treated_data = pd.read_stata(treated_file_path)

# Prepare data for analysis
data['after'] = (data['year'] >= 2009).astype(int)
combined_data = data.merge(treated_data, on='country', how='left')
combined_data['treated'] = combined_data['treated'].fillna(0).astype(int)
combined_data['T_it'] = ((combined_data['treated'] == 1) & (combined_data['after'] == 1)).astype(int)

# Rename and select necessary columns
prepared_data = combined_data.rename(columns={'year': 't', 'country': 'i', 'gdppc': 'Y_it'})
prepared_data = prepared_data[['t', 'i', 'Y_it', 'T_it', 'treated', 'after', 'exports', 'imports', 'labor', 'trade',
       'ln_gdppc', 'ln_labor', 'ln_trade']]
prepared_data = prepared_data.set_index(['i', 't'])

# Define dependent and explanatory variables
exog_vars = ['T_it', 'exports', 'imports', 'labor']
exog = prepared_data[exog_vars]

# (1) Fixed Effects, With and Without Clustering
mod_fixed = PanelOLS(prepared_data['Y_it'], exog, entity_effects=True, drop_absorbed=True)
results_fixed_no_cluster = mod_fixed.fit()
results_fixed_cluster = mod_fixed.fit(cov_type="clustered", cluster_entity=True)

# (2) Fixed + Time Effects, With and Without Clustering
mod_fixed_time = PanelOLS(prepared_data['Y_it'], exog, entity_effects=True, time_effects=True, drop_absorbed=True)
results_fixed_time_no_cluster = mod_fixed_time.fit()
results_fixed_time_cluster = mod_fixed_time.fit(cov_type="clustered", cluster_entity=True, cluster_time=True)

# Compare results
results_table = compare({
    "FE": results_fixed_no_cluster,
    "FE (Clustered)": results_fixed_cluster,
    "TWFE": results_fixed_time_no_cluster,
    "TWFE (Clustered)": results_fixed_time_cluster
})

# Display results
print(results_table)
```

#### Matched TWFE

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import NearestNeighbors
from linearmodels.panel import PanelOLS, compare

# Load data
file_path = 'data/WDI.dta'
treated_file_path = 'data/Treated.dta'

data = pd.read_stata(file_path)
treated_data = pd.read_stata(treated_file_path)

# Prepare data for analysis
data['after'] = (data['year'] >= 2009).astype(int)
combined_data = data.merge(treated_data, on='country', how='left')
combined_data['treated'] = combined_data['treated'].fillna(0).astype(int)
combined_data['T_it'] = ((combined_data['treated'] == 1) & (combined_data['after'] == 1)).astype(int)

# Step 1: Propensity Score Matching
covariates = ['exports', 'imports', 'labor', 'trade', 'ln_labor', 'ln_trade']
X = combined_data[covariates]
y = combined_data['treated']

# Logistic regression for propensity scores
logit = LogisticRegression(max_iter=1000)
combined_data['propensity_score'] = logit.fit(X, y).predict_proba(X)[:, 1]

# Matching treated and untreated units based on propensity scores
treated = combined_data[combined_data['treated'] == 1]
untreated = combined_data[combined_data['treated'] == 0]

nn = NearestNeighbors(n_neighbors=1)
nn.fit(untreated[['propensity_score']])
distances, indices = nn.kneighbors(treated[['propensity_score']])
matched_untreated = untreated.iloc[indices.flatten()]

# Combine treated and matched untreated data
matched_data = pd.concat([treated, matched_untreated])

# Step 2: Difference-in-Differences (DID)
matched_data = matched_data.rename(columns={'year': 't', 'country': 'i', 'gdppc': 'Y_it'})
matched_data = matched_data[['t', 'i', 'Y_it', 'T_it', 'treated', 'after'] + covariates]
matched_data = matched_data.set_index(['i', 't'])

# Define dependent and explanatory variables
exog_vars = ['T_it']
exog = matched_data[exog_vars]

# (1) Fixed Effects, With and Without Clustering
mod_fixed = PanelOLS(matched_data['Y_it'], exog, entity_effects=True, drop_absorbed=True)
results_fixed_no_cluster = mod_fixed.fit()
results_fixed_cluster = mod_fixed.fit(cov_type="clustered", cluster_entity=True)

# (2) Fixed + Time Effects, With and Without Clustering
mod_fixed_time = PanelOLS(matched_data['Y_it'], exog, entity_effects=True, time_effects=True, drop_absorbed=True)
results_fixed_time_no_cluster = mod_fixed_time.fit()
results_fixed_time_cluster = mod_fixed_time.fit(cov_type="clustered", cluster_entity=True, cluster_time=True)

# Compare results
results_table = compare({
    "FE": results_fixed_no_cluster,
    "FE (Clustered)": results_fixed_cluster,
    "TWFE": results_fixed_time_no_cluster,
    "TWFE (Clustered)": results_fixed_time_cluster
})

# Display results
print(results_table)

import matplotlib.pyplot as plt
import seaborn as sns

# KDE plot of propensity scores before matching
plt.figure(figsize=(10, 6))
sns.kdeplot(data=treated['propensity_score'], label='Treated (Before Matching)', fill=True, alpha=0.5)
sns.kdeplot(data=untreated['propensity_score'], label='Untreated (Before Matching)', fill=True, alpha=0.5)
plt.title('Propensity Score Distribution Before Matching')
plt.xlabel('Propensity Score')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

# KDE plot of propensity scores after matching
plt.figure(figsize=(10, 6))
sns.kdeplot(data=treated['propensity_score'], label='Treated (After Matching)', fill=True, alpha=0.5)
sns.kdeplot(data=matched_untreated['propensity_score'], label='Matched Untreated (After Matching)', fill=True, alpha=0.5)
plt.title('Propensity Score Distribution After Matching')
plt.xlabel('Propensity Score')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

### Organ Donation

```python
import pyfixest as pf
from causaldata import organ_donations
import pandas as pd
import matplotlib.pyplot as plt

df = organ_donations.load_pandas().data

df["California"] = df["State"] == "California"
df["After"] = df["Quarter_Num"] > 3
df["Treated"] = 1 * (df["California"] & df["After"])

df = df.rename(columns={'State': 'i', 'Quarter_Num': 't', 'Rate': 'Y_it', 'Treated': 'T_it'})

df['ever_treated'] = df['California'].astype(int)
df['time_to_treat'] = df['t'] - 4
df.loc[df['ever_treated'] == 0, 'time_to_treat'] = None

california_data = df[df['i'] == 'California']
others_data = df[df['i'] != 'California'].groupby('t')['Y_it'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(california_data['t'], california_data['Y_it'], label='California', marker='o')
plt.plot(others_data['t'], others_data['Y_it'], label='Other States (Average)', marker='s')
plt.axvline(x=4, color='red', linestyle='--', label='Treatment Begins (t = 4)')
plt.title('Y_it Over Time: California vs. Other States')
plt.xlabel('Quarter (t)')
plt.ylabel('Y_it (Rate)')
plt.legend()
plt.grid(True)
plt.show()

display(df.head())
```
