### Replicating Statsmodels

```python
import math
from numpy.random import normal, seed
from numpy import dot, sqrt, diagonal, where
from numpy.linalg import inv
import numpy as np
from scipy.stats import t, norm, f

seed(1)
n, k = 100, 10
β = normal(10, 1, (k, 1))
X = normal(0, 1, (n, k))
ε = normal(0, 1, (n, 1))
y = dot(X, β) + ε
X[:, 0:3] = normal(0, 1, (n, 3))

A = inv(dot(X.T, X))
b = dot(A, dot(X.T, y))
ŷ = dot(X, b)
e = y - ŷ
RSS = float(dot(e.T, e))
ESS = float(dot(ŷ.T, ŷ))
TSS = float(dot(y.T, y))
s = sqrt(RSS / (n - k))
b_se = diagonal(s * sqrt(where(A < 0, 0, A))).reshape(-1, 1)
t_stats = b.flatten() / b_se.flatten()

df = n - k
p_values = [2 * (1 - t.cdf(abs(t_stat), df)) for t_stat in t_stats]
α = 0.05
c = t.ppf(1 - α / 2, df)
lower_conf = b.flatten() - c * b_se.flatten()
upper_conf = b.flatten() + c * b_se.flatten()

R2_uncentered = ESS / TSS
Adj_R2_uncentered = 1 - (1 - R2_uncentered) * (n / (n - k))
F_statistic = (ESS / k) / (RSS / (n - k))
log_likelihood = -0.5 * n * (math.log(2 * math.pi) + math.log(RSS / n) + 1)
AIC = 2 * k - 2 * log_likelihood
BIC = k * math.log(n) - 2 * log_likelihood

e_mean = sum(e) / len(e)
e_variance = sum((e - e_mean) ** 2) / len(e)
e_std = sqrt(e_variance)
e_skewness = (sum((e - e_mean) ** 3) / n) / e_std**3
e_kurtosis = (sum((e - e_mean) ** 4) / n) / e_std**4
omnibus_stat = n * (e_skewness**2 / 6 + (e_kurtosis - 3)**2 / 24)
omnibus_p = 2 * (1 - f.cdf(sqrt(omnibus_stat)))
dw_stat = sum((e[i] - e[i - 1]) ** 2 for i in range(1, len(e))) / sum(e[i] ** 2 for i in range(len(e)))
jb_stat = n * (e_skewness**2 / 6 + (e_kurtosis - 3)**2 / 24)
jb_p = 1 - norm.cdf(jb_stat)
condition_number = max(np.linalg.svd(X, compute_uv=False)) / min(np.linalg.svd(X, compute_uv=False))

print(f"""
                                 OLS Regression Results                                
=======================================================================================
Dep. Variable:                      y   R-squared (uncentered):                {R2_uncentered:.3f}
Model:                            OLS   Adj. R-squared (uncentered):           {Adj_R2_uncentered:.3f}
Method:                 Least Squares   F-statistic:                           {F_statistic:.2f}
Date:                {pd.Timestamp.now().strftime('%a, %d %b %Y')}   Prob (F-statistic):               {1 - f.cdf(F_statistic, k, n - k):.2e}
Time:                        {pd.Timestamp.now().strftime('%H:%M:%S')}   Log-Likelihood:                        {log_likelihood:.2f}
No. Observations:                 {n}   AIC:                                   {AIC:.1f}
Df Residuals:                      {n - k}   BIC:                                   {BIC:.1f}
Df Model:                          {k}                                                  
Covariance Type:            nonrobust                                                  
==============================================================================
                 coef    std err          t      P>|t|      [0.025      0.975]
------------------------------------------------------------------------------
""")
for i in range(k):
    print(f"x{i+1:<2} {b.flatten()[i]:>10.4f} {b_se.flatten()[i]:>10.4f} {t_stats[i]:>10.4f} "
          f"{p_values[i]:>10.4f} {lower_conf[i]:>10.4f} {upper_conf[i]:>10.4f}")
print(f"""==============================================================================
Omnibus:                        {float(omnibus_stat):.3f}   Durbin-Watson:                   {float(dw_stat):.3f}
Prob(Omnibus):                  {float(omnibus_p):.3f}   Jarque-Bera (JB):                {float(jb_stat):.3f}
Skew:                           {float(e_skewness):.3f}   Prob(JB):                        {float(jb_p):.3f}
Kurtosis:                       {float(e_kurtosis):.3f}   Cond. No.                         {float(condition_number):.2f}
==============================================================================

Notes:
[1] R² is computed without centering (uncentered) since the model does not contain a constant.
[2] Standard Errors assume that the covariance matrix of the errors is correctly specified.
""")
```

```python
from statsmodels.api import OLS
model = OLS(y, X)
result = model.fit()
print(result.summary())
```

### Multivariate Regression - with interactions, HAC stderrs, cluster stderrs

```python
# Read in data
dt = Mroz.load_pandas().data
print(dt.dtypes)
dt = dt.query("lfp") # working women
dt.loc[:, "earn"] = dt["lwg"].apply("exp") # unlogged earnings
fit = pf.feols(fml="lwg ~ csw(inc, wc, k5)", data=dt, vcov="iid")
pf.etable(fit)
```

```python
res = restaurant_inspections.load_pandas().data
res.inspection_score = res.inspection_score.astype(float)
res.NumberofLocations = res.NumberofLocations.astype(float)
print(res.dtypes)
fit = pf.feols(fml="inspection_score ~ NumberofLocations", data=res)
pf.etable([fit])
```

```python
df = restaurant_inspections.load_pandas().data
fit1 = pf.feols(
    fml="inspection_score ~ NumberofLocations + I(NumberofLocations^2) + Year", data=df
)
fit2 = pf.feols(fml="inspection_score ~ NumberofLocations*Weekend + Year", data=df)
pf.etable([fit1, fit2])
```

```python
pf.feols(fml="inspection_score ~ Year + Weekend", data=df, vcov="HC3").summary()
```

```python
pf.feols(fml="inspection_score ~ Year + Weekend", data=df, vcov={"CRV1": "Weekend"}).tidy()
```

### Fixed and Two way fixed effects

```python
gm = gapminder.load_pandas().data
gm["logGDPpercap"] = gm["gdpPercap"].apply("log")
display(gm)
fit = pf.feols(fml="lifeExp ~ C(country) + np.log(gdpPercap)", data=gm)
fit.tidy().head()
```

```python
# Set our individual and time (index) for our data
fit = pf.feols(fml="lifeExp ~ np.log(gdpPercap) | country + year", data=gm)
fit.summary()
```

```python
od = organ_donations.load_pandas().data

# Create Treatment Variable
od["California"] = od["State"] == "California"
od["After"] = od["Quarter_Num"] > 3
od["Treated"] = 1 * (od["California"] & od["After"])

did = pf.feols(fml="Rate ~ Treated | State + Quarter", data=od)
did.summary()
```
